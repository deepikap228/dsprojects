```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2007,1,1)
```

```python
#Bank of America
BAC = data.DataReader('BAC','robinhood',start,end)
```

```python
# Get the general idea of the dataset
BAC.head()
```

```python
# CitiGroup
C = data.DataReader("C", 'robinhood', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'robinhood', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'robinhood', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'robinhood', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'robinhood', start, end)
```

```python
tickers=['BAC','C','GS','JPM','MS','WFC']

tickers
```

```python
BAC.reset_index(drop=True,  
C.reset_index(drop=True, inplace=True)
GS.reset_index(drop=True, inplace=True)
JPM.reset_index(drop=True, inplace=True)
MS.reset_index(drop=True, inplace=True)
WFC.reset_index(drop=True, inplace=True)

bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
for tick in tickers:
    print(tick,bank_stocks[tick]['close_price'].max())
```

```python
bank_stocks.xs(key='close_price',axis=1,level='Stock Info').max()
```

```python
returns=pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+'Return']=bank_stocks[tick]['close_price'].pct_change()
```
